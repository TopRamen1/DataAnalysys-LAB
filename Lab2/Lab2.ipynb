## Excercise 3 - Constraints on parameters

### Unconstrained parameters

```python
model_gm1 = CmdStanModel(stan_file='./stan_files/code_4.stan')
out_gamma1 = model_gm1.sample(output_dir='samples',iter_sampling=6000,iter_warmup=1000, seed=4838282)

```

```python
out_gamma1.diagnose()
```

```python
N=500
xs = np.linspace(0,8,N)
pdfs = stats.gamma.pdf(xs, 1.25, scale = 1 / 1.25) 

plt.plot(xs, pdfs, linewidth=2)

plt.hist(out_gamma1.stan_variable('theta'), bins=160)

plt.gca().set_xlabel("theta")
plt.gca().set_ylabel("Probability Density Function")
plt.show()
```

### Constrained parameter

```python
model_gm2 = CmdStanModel(stan_file='./stan_files/code_5.stan')
out_gamma2 = model_gm2.sample(output_dir='samples',iter_sampling=6000,iter_warmup=1000, seed=4838282)
out_gamma2.diagnose()
```

```python
N=500
xs = np.linspace(0,8,N)
pdfs = stats.gamma.pdf(xs, 1.25, scale = 1 / 1.25) 

plt.plot(xs, pdfs, linewidth=2)

plt.hist(out_gamma2.stan_variable('theta'), bins=160)

plt.gca().set_xlabel("theta")
plt.gca().set_ylabel("Probability Density Function")
plt.show()
```

## Excercise 5 - different methods of defining models


```python
model_samp_st = CmdStanModel(stan_file='./stan_files/code_7.stan')
model_log_target = CmdStanModel(stan_file='./stan_files/code_8.stan')
model_log_target_ind = CmdStanModel(stan_file='./stan_files/code_9.stan')
data = {'N': F}
seed = 13072000 #integer, your date of birth in the DDMMYYYY format without leading zero (or if you are GPRD weary, use any other date you wish)
result_1 = model_samp_st.sample(data=data,seed=seed)
result_2 = model_log_target.sample(data=data,seed=seed)
result_3 = model_log_target_ind.sample(data=data,seed=seed)

```
